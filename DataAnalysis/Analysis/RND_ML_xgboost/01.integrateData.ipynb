```python
# 날짜관련 뽑아내기 

# '주중주말',
# '년도',
# '월',
# '주차',
# '요일',
# '공휴일',
test = Service.dayToIntConvert(integrated_df,'수송일자') # 5분 34초
print(test.columns)
test.drop(['Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'], inplace=True,axis=1)
test.head()
```

```python

test.columns
```

```python
test_for_merge = test
dt_for_merge = dispatch_table.drop('호선',axis=1)
```

```python
dt_for_merge
change_colName =[f'{i:02d}배차' for i in range(5,25)]
for idx,col in enumerate(dt_for_merge.columns[2:]):
    print(idx,col,change_colName[idx])
    dt_for_merge.rename({col:change_colName[idx]},inplace=True,axis=1)
dt_for_merge
```

```python
new_code =[]
for i in test_for_merge['역사코드']:
    # print(str(type(i)),str(type("str")))
    if str(type(i)) == str(type("")):
        # print("String")
        if i.strip() == '':
            new_code.append(0)
        else:
            new_code.append(int(i.strip()))
    else:
        new_code.append(i)
    
len(set(new_code))
test_for_merge['역사코드']=new_code
```

```python
test_for_merge['역사코드'].
dt_for_merge['역사코드']
```

```python
result =pd.merge(test_for_merge,dt_for_merge,on=['역사코드','주중주말'])
print(result.columns)
print(result.info())
result.head(3)
```

```python
result.to_csv(os.path.join(data_Folder,'finalTable_18to24.csv'),index=None)
```
